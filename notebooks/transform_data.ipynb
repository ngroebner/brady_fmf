```python
continuous_nodal_waveforms = pd.read_csv("../data/raw/nodal_seismometer_metadata/Nodal_continuous_file_list.csv")
```

```python
continuous_nodal_waveforms
```

```python
continuous_nodal_waveforms['Directory'] = continuous_nodal_waveforms['Directory'].apply(lambda x: str(x).replace(".","").replace(":","/"))
continuous_nodal_waveforms.head()
```

```python
continuous_nodal_waveforms['file_name'] = continuous_nodal_waveforms['file_name'].apply(lambda x: str(x).replace("*",""))
continuous_nodal_waveforms["Directory"] = continuous_nodal_waveforms["Directory"].apply(lambda x: x.replace("NODAL_SAC","nodal_sac"))
continuous_nodal_waveforms.head()
```

```python
waveforms = pd.DataFrame(continuous_nodal_waveforms['Directory'] + continuous_nodal_waveforms['file_name'])
```

```python
waveforms.columns = ["filename"]
waveforms.head()
```

### Filenames are wrong in the original file!!! Need to add 1000 to the number before channel designator

```python
waveforms = waveforms[waveforms["filename"].apply(lambda x: len(x)>=45)]
```

```python
waveforms["station"] = waveforms["filename"].apply(lambda x: x.split("/")[2])
waveforms["channel"] = waveforms["filename"].apply(lambda x: x.split(".")[2])

# add 100 to number (somoe sort of timestamp I think)
def convert_name(x):
    timestamp_old = x.split("/")[3].split(".")[0]
    timestamp_new = str(int(timestamp_old) + 1000)
    x = x.replace(timestamp_old, timestamp_new)
    return x

waveforms["filename"] = waveforms["filename"].apply(convert_name)
```

```python
waveforms['date'] = waveforms["filename"].apply(lambda x: x.split("/")[3][:8]).apply(pd.to_datetime)
```

```python
waveforms
```

```python
waveforms.to_pickle("../data/processed/continuous_waveform_filenames.pkl")
```
